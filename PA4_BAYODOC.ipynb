# 1. Create the following data frames based on the format provided:
Example: Vis = [“Name”, “Gender”, “Track”, “Math<70”]; hometown is constant as Visayas
Output:


- Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as Instrumentation and hometown Luzon
- Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female

```python
# Imports the pandas library with an alias pd
import pandas as pd

df = pd.read_excel('board2.xlsx')
df
```

### Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as Instrumentation and hometown Luzon

```python
# The conditions are as follows: 
# 1. The track should be 'Instrumentation'
# 2. Hometown should be in 'Luzon'
# 3. Their grade in 'Electronics' > 70

Instru = df.loc[ (df['Electronics'] > 70) & 
    (df['Track'] == 'Instrumentation') & 
    (df['Hometown'] == 'Luzon'), 
    [ 'Name', 'GEAS', 'Electronics']  ] 

Instru
```

### Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female

```python
# Computing for the average in math, electronics, GEAS, and communication

GWA = (df["Math"]+ df["Electronics"]+ df["GEAS"]+ df["Communication"])/4
GWA
```

```python
# Filters the condition such that their hometown must be in "Mindanao" and they must be 'female'

Mindy = df.loc[ (df["Hometown"] == "Mindanao") & (df["Gender"] == "Female"), ["Name", "Track", "Electronics"] ]
Mindy
```

```python
# Insert the data series, GWA, in Mindy
# Then, it filters the rows with an average greater than or equal to 55


Mindy['Average'] = GWA
Mindy[ Mindy['Average'] >= 55 ]
```

### 2. Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown contributes to a higher average score?



```python
# importing important libraries

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure( figsize=(15, 5) )

# The code below plots the effect of the students' track to their average score
plt.subplot(1, 3, 1)
sns.boxplot(x = 'Track', y = GWA, data = df)
plt.title( 'Average Score by Track' )
plt.xticks( rotation = 45 )
```

```python
# The code below plots the effect of the students' gender to their average score
plt.subplot( 1, 3, 2 )
sns.boxplot( x='Gender', y=GWA, data=df )
plt.title( 'Average Score by Gender' )
```

```python
# The code below plots the effect of the student's hometown to their average score
plt.subplot( 1, 3, 3 ) 
sns.boxplot( x ='Hometown', y = GWA, data = df )
plt.title( 'Average Score by Hometown' )
plt.xticks( rotation = 45 )

plt.tight_layout()
plt.show()
```

# Thus, the chosen track in college, gender, or hometown does not contribute to a higher average score.
